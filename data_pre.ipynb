# 여러개의 json(ZIP파일) 하나로 합침

```python
import zipfile
import json
import pandas as pd

def extract_json_from_zip(zip_path):
    # 빈 DataFrame 생성
    df_all = pd.DataFrame()

    # ZIP 파일 처리
    with zipfile.ZipFile(zip_path, 'r') as zip_file:
        # ZIP 파일 내 JSON 파일 목록 가져오기
        json_files = [file for file in zip_file.namelist() if file.endswith('.json')]

        # 각 JSON 파일 처리
        for json_file in json_files:
            with zip_file.open(json_file) as file:
                # JSON 파일 읽기
                json_data = json.load(file)

                # 필요한 데이터를 DataFrame 형식으로 변환 (예: 'dataSet'만 추출)
                data = json_data.get('dataSet', {})
                if data:
                    df = pd.json_normalize(data)
                    df_all = pd.concat([df_all, df], ignore_index=True)

    return df_all

# ZIP 파일 경로들
zip_paths = [
    "Downloads/NLP/Training/TL_05.ICT_Female_New.zip",
    "Downloads/NLP/Training/TL_05.ICT_Male_New.zip",
    "Downloads/NLP/Training/TL_05.ICT_Male_Experienced.zip",
    "Downloads/NLP/Training/TL_05.ICT_Female_Experienced.zip",
]

# 각 ZIP 파일 처리
df1 = extract_json_from_zip(zip_paths[0])
df2 = extract_json_from_zip(zip_paths[1])
df3 = extract_json_from_zip(zip_paths[2])
df4 = extract_json_from_zip(zip_paths[3])

# 모든 DataFrame을 합침
df_all = pd.concat([df1, df2, df3, df4], ignore_index=True)|

# 결과 출력
print(df_all.head())
```

```python
# 'answer.emotion'을 리스트로 변환 (JSON 방식으로 처리)
df_all['answer.emotion'] = df_all['answer.emotion'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)

# 감정 데이터가 담긴 각 항목을 분리하여 새로운 컬럼에 저장 (리스트 형태로 반환)
df_all['emotion_text'] = df_all['answer.emotion'].apply(lambda x: [f"'{item['text']}'" for item in x] if isinstance(x, list) else [])
df_all['emotion_expression'] = df_all['answer.emotion'].apply(lambda x: [f"'{item['expression']}'" for item in x] if isinstance(x, list) else [])
df_all['emotion_category'] = df_all['answer.emotion'].apply(lambda x: [f"'{item['category']}'" for item in x] if isinstance(x, list) else [])

# 'answer.intent' 컬럼을 리스트로 변환 (JSON 형식으로 처리)
df_all['answer.intent'] = df_all['answer.intent'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)

# 각 항목의 'text', 'expression', 'category'를 별도 컬럼으로 분리
# 각 항목을 ' '로 감싸서 리스트로 저장
df_all['intent_text'] = df_all['answer.intent'].apply( lambda x: [f"'{item['text']}'" for item in x] if isinstance(x, list) else [])
df_all['intent_expression'] = df_all['answer.intent'].apply(lambda x: [f"'{item['expression']}'" for item in x] if isinstance(x, list) else [])
df_all['intent_category'] = df_all['answer.intent'].apply(lambda x: [f"'{item['category']}'" for item in x] if isinstance(x, list) else [])

df_all.head(5)
```
